# Tuple

A Tuple is a built-in data type in Python used to store multiple values in a single variable.

👉 Tuple is similar to a List but with one major difference:

List	Tuple
Mutable (can change)	Immutable (cannot change)
Uses []	Uses ()
Slower	Faster
More memory	Less memory

Faster than List

Tuple operations are generally faster because tuples are immutable.

```python
t = (1,2,3)
print(t)
print(type(t))
```

```python
#ordered and mixed data type and duplication 
t = (1,.14,'dinal',True,1,1,2,3,3)
print(t)
```

## Different ways

```python
# empty tuple 
t = ()
print(t)
print(type(t))
```

```python
# python treat comma separated variable as tuple
t = 1,2,3,4,5
print(type(t))
```

```python
#using constructer

t = tuple([1,2,3,4])
print(t)
```

## single element tuple

```python
t = (3)#incorrect
print(type(t))
```

```python
t = (3,)
print(type(t))
```

## Accessing and slicing

```python
t = ("apple", "banana", "mango","banana","watermelon","kiwi","peach","orange","strawberry")

print(t[0])
print(t[1])
print(t[2])
```

```python
t[-1]
```

```python
t[::2]
```

```python
t[1:5]
```

## Immutability of Tuple

Data safety
Faster execution
Prevent accidental changes

```python
t = (1, 2, 3)

t[0] = 100
```

```python
# count

t = (1,2,2,2,3,4,4,5,5)
t.count(2)
```

```python
t.index(2)#forst apperaning
```

```python
#packing
t = 1,3,2
print(t)
```

```python
#unpacking
t= (10,20,30)
a,b,c = t
```

```python
print(a)
print(b)
print(c)
```

```python
#swap
a = 10
b = 20

a, b = b, a

print(a)
print(b)
```

```python
t = (1, 2, 3, 4, 5)

a, *b, c = t

print(a)
print(b)
print(c)
```

```python
#loop
t = ("apple", "banana", "mango")

for item in t:
    print(item)
```

```python
t = (10, 20, 30)

for i in range(len(t)):
    print(i, t[i])
```

```python
t = ((1, 2), (3, 4), (5, 6))

print(t)
```
