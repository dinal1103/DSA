# Function

A function is a reusable block of code that performs a specific task.
Instead of writing the same code again and again, we can place it inside a function and call it whenever needed.

Functions help in:

✅ Reducing code repetition
✅ Making code cleaner
✅ Improving readability
✅ Breaking large programs into smaller parts
✅ Easier debugging and maintenance

```python
def greet():
    print("hello")
```

```python
greet()
```

```python
# with parameters
def greet(name):
    print("hello " , name)
```

```python
greet("dinal")
```

```python
greet("world")
```

```python
#multiple parameter
def add(a,b,c):
    return(a+b+c)
```

```python
add(1,2,3)
```

```python
add(1,2)
```

```python
# default parameter

def greet(name = "guest"):
    print("hello ",name)
```

```python
greet()
greet("dinal")
```

```python
# positinal argument
# Values are assigned according to position.
def student(name, age):
    print(name)
    print(age)

student("Aman", 20)
```

name - aman and age - 0

```python
# keyword argument
# Arguments passed using parameter names.
def student(name, age):
    print(name)
    print(age)

student(age=20, name="Aman")
```

Variable Length Arguments

Sometimes we don't know how many arguments user will pass.

| Symbol     | Meaning                       |
| ---------- | ----------------------------- |
| `*args`    | Multiple positional arguments |
| `**kwargs` | Multiple keyword arguments    |


*args stores multiple values in tuple.

```python
def numbers(*nums):
    print(nums)

numbers(1, 2, 3)
```

```python
def total(*values):
    print(sum(values))

total(10, 20, 30)
```

**kwargs stores multiple keyword arguments in dictionary.

```python
def info(**data):
    print(data)

info(name="Dinal", age=20)
```

```python
def student(**details):
    for key, value in details.items():
        print(key, value)

student(name="Rahul", age=21)
```

```python
def employee(**emp):
    print(emp["name"])

employee(name="Amit", salary=50000)
```

```python
#recursive funstion
def countdown(n):

    if n == 0:
        return

    print(n)

    countdown(n - 1)

countdown(5)
```
